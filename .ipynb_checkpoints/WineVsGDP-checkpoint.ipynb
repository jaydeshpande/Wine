```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
def runGDP(countries):
    gdpbycntr = pd.read_csv('GDP.csv')
    print ("Total dataset length: {}",format(len(gdpbycntr)))
    cntrcode = pd.read_csv('CountryCodes.csv')
    gdpDat = gdpbycntr.loc[gdpbycntr['Name'].isin(cntrcode['Name'])]
    print ("Total dataset length: {}",format(len(gdpDat)))
    gdpDat.drop(['CountryCode', 'Indicator', 'Indicator Code'], axis = 1, inplace=True)
    yrs = gdpDat.columns.values[3:]
    datToPlot = gdpDat.loc[gdpDat['Name'].isin(countries)].set_index('Name').T
    datToPlot['Year'] = datToPlot.index.values
    #print (datToPlot.head())
    #datToPlot.to_csv('test.csv')
    buildLinePlot(datToPlot)

def buildLinePlot(dataset):
    dataset = dataset.melt('Year', var_name='Countries',  value_name='GDP')
    ax = sns.lineplot(x='Year', y='GDP',hue='Countries',  data=dataset)
    plt.show()
```

```python
runGDP(['France', 'Germany'])
```

```python
winedat = pd.read_excel('beerwinedat.xlsx', sheetname='Vol_wine_kl_alc').T
```

```python
beerdat = pd.read_excel('beerwinedat.xlsx', sheetname='Vol_beer_kl_alc').T
```

```python
winedat = winedat[winedat['Decade'] > 10]
beerdat = beerdat[beerdat['Decade'] > 10]
```

```python
winedat.replace('na', 0, inplace=True)
beerdat.replace('na', 0, inplace=True)
```

```python
def runWine(countries,source):
    fig, ax = plt.subplots(figsize=(8, 6))
    for i in countries: 
        ax.scatter(np.arange(1, 1+len(source.Decade.values)), source[i], label=i)
    ax.set_xticks(np.arange(1, 1+len(source.Decade.values)))
    ax.set_xticklabels(source.index.values)
    ax.set_ylabel('Volume of alcohol in spirit [kL]')
    ax.set_xlabel('Decades')
    plt.legend()
    plt.show()
```

```python
runWine(['France', 'Spain', 'Italy', 'Portugal'], source=winedat)
```

```python
runWine(['France', 'Spain', 'Italy', 'Portugal'], source=beerdat)
```

```python
runWine(['USA'], source=beerdat)
```
